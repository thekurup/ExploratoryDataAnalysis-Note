```python
df=pd.read_excel('E:\LEARNING AI\EDA/Country-Code.xlsx')
```

```python
df
```

```python
data=pd.merge(dff,df,on='Country Code',how='left')
```

```python
data
```

```python
#check most transactions based country
```

```python
x=data['Country'].value_counts().index
```

```python
y=data['Country'].value_counts().values
```

```python
x
```

```python
y
```

```python
#plotiing in pie
```

```python
plt.pie(y[:3],labels=x[:3],radius=1.5,autopct='%1.2f%%',explode=[0.1,0.1,0.1],shadow=True)
plt.show()
```

```python
#check most rated values 
```

```python
data['Aggregate rating'].value_counts().sort_values(ascending=False)
```

```python
ratings=data.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0 : 'rated value'})
```

```python
ratings
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating', y='rated value',data=ratings,hue='Rating color')
plt.show()
```

```python
sns.countplot(x="Rating color",data=ratings)
```
